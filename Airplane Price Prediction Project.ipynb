```python
## Checking outliers using Z-score techniques
import numpy as np
from scipy.stats import zscore

# Identify numeric columns first
numeric_columns = plane.select_dtypes(include=[np.number]).columns

# Calculate z-scores for each column
plane[numeric_columns] = plane[numeric_columns].apply(zscore)

# Identify outliers based on Z-scores (e.g., considering Z-score > 3 or < -3 as outliers)
outliers = (abs(plane[numeric_columns]) > 3).any(axis=1)
outliers_df = plane[outliers]
outliers_df
```

```python
# label_Encoding

from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
plane1 = plane
#plane['Model_Name'] = label_encoder.fit_transform(plane['Model_Name'])
```

```python
plane1['Landing_Over_50ft'] = plane1['Landing_Over_50ft'].str.replace(',', '').astype(float)
plane1['Empty_Weight_lbs'] = plane1['Empty_Weight_lbs'].str.replace(',', '').astype(float)
```

```python
plane1['Engine_Type_Piston'] = plane1['Engine_Type_Piston'].astype('int')
plane1['Engine_Type_Propjet'] = plane1['Engine_Type_Propjet'].astype('int')
plane1['Engine_Type_piston'] = plane1['Engine_Type_piston'].astype('int')
plane1['Engine_Type_Jet'] = plane1['Engine_Type_Jet'].astype('int')
```

```python
# Feature Selection techniques - Correlation

plane1.corr().sum()
```

```python
plt.figure(figsize = (10, 6))
sns.heatmap(plane1.corr(), annot = True)
```

```python
correlation_matrix = plane1.corr()
```

```python
target_correlation_threshold = 0.5  # Adjusting the threshold
target_correlation = correlation_matrix['Price'].abs()
selected_features = target_correlation[target_correlation > target_correlation_threshold].index.tolist()
```

```python
feature_correlation_threshold = 0.8  # Adjusting the threshold 
correlated_features = set()

for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > feature_correlation_threshold:
            colname = correlation_matrix.columns[i]
            correlated_features.add(colname)

df_filtered = plane1.drop(columns=correlated_features)
```

```python
## Spliting the data

X = df_filtered.drop(['Length_ft/in', 'Wing_Span_ft/in'], axis = 1)
y = plane1['Price']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=100) 
```

```python
## Modeling

from sklearn.ensemble import RandomForestRegressor
random_reg = RandomForestRegressor()
random_reg.fit(X_train, y_train)
```

```python
y_pred3 = random_reg.predict(X_test)
```

```python
random_reg.score(X_test, y_test)
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred3))
print('MSE:', metrics.mean_squared_error(y_test, y_pred3))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred3)))
```
